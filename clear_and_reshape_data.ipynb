```python
# Выявляем англоязычные тексты
from langdetect import detect, DetectorFactory
DetectorFactory.seed = 0
detect_lang = []
for text in book['Description']:
    try:
        detect_lang.append(detect(text))
    except:
        print(len(detect_lang), text)
        detect_lang.append('no')
```

```python
def load_stop_words():
    with open("source/language.txt", "r") as file:
        lines = file.readlines()
        language = [line.strip().split()[0] for line in lines]

        stop_words = []
        for lang in language:
            stop_words += stopwords.words(lang)
        return stop_words + ['.', ',', "''", '--', '-', "``", '“', '”', '(', ')', ':', ';', '!', '?', '/', '%', '&', '@', '#', '…', '...', '’', '....']
```
