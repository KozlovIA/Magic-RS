```python
import pandas as pd
```

```python
from os import listdir
from os.path import isfile, join
onlyfiles = [f for f in listdir('archive') if isfile(join('archive', f))] # поиск всех файлов с данными
```

```python
book = pd.DataFrame(columns=['Id', 'Name', 'Authors', 'ISBN', 'Rating', 'PublishYear',
       'PublishMonth', 'PublishDay', 'Publisher', 'RatingDist5', 'RatingDist4',
       'RatingDist3', 'RatingDist2', 'RatingDist1', 'RatingDistTotal',
       'CountsOfReview', 'Language', 'pagesNumber', 'Description',
       'Count of text reviews'])
for file in onlyfiles:
    temp = pd.read_csv('archive/' + file)
    if 'Description' in temp.columns:
        book = pd.concat([book, temp])
```

```python
book
```

## Поиск и запись в файл датасета с запясями, где есть описание

```python
book = book[~book['Description'].isna()] # а потом это to_csv
book = book[~book['Name'].duplicated()]     # Убираем дубликаты книг по названию
book.to_csv('book_with_description.csv')
```

# Работа с данными о юзерах

```python
user = pd.DataFrame(columns=['ID', 'Name', 'Rating'])
for file in onlyfiles:
    if 'user' in file:
        user = pd.concat([user, pd.read_csv('archive/' + file)])
```

```python
df = user.join(book[['Name', 'Description']].set_index('Name'), how='inner', on='Name')
```
